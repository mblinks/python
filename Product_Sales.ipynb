## Visualizations

```python
# Number of customers for each sales method
colors = sns.color_palette("pastel")
customer_count.plot(kind='bar', rot=0, xlabel='Sales Method', ylabel='Number of Customers', title='Number of Customers for Each Sales Method', color=colors)
plt.savefig('Task 1.png')
```

```python
# Create a histogram to visualize the distribution of revenue
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='revenue', bins=10, kde=False)
plt.title('Distribution of Overall Revenue')
plt.xlabel('Revenue')
plt.ylabel('Count')

# Show the plot
plt.savefig('Task 6.png')
```

```python
# Group the data by sales method and count unique customer_id values
customer_count_by_method = df.groupby('sales_method')['customer_id'].nunique()

# Display the total number of customers for each sales method
print(customer_count_by_method)
```

```python
# Create a figure for revenue spread by sales method
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='sales_method', y='revenue')
plt.title('Revenue Spread by Sales Method')
plt.xlabel('Sales Method')
plt.ylabel('Revenue')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the second plot
plt.savefig('Task 3.png')
```

```python
# Group the data by 'sales_method' and 'week' and sum the revenue
revenue_by_method_week = df.groupby(['sales_method', 'week'])['revenue'].sum().unstack().T

# Create a bar plot to visualize revenue per sales method per week
plt.figure(figsize=(18, 9))
revenue_by_method_week.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Revenue per Sales Method per Week')
plt.xlabel('Week')
plt.ylabel('Revenue')
plt.legend(title='Sales Method', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Relationship between sales method and years as customers
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='sales_method', y='years_as_customer', palette='Set2')
plt.title('Relationship between Sales Method and Years as Customer')
plt.xlabel('Sales Method')
plt.ylabel('Years as Customer')

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
